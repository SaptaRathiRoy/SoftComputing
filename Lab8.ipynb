# Program 14A

```python
import random
import matplotlib.pyplot as plt
```

```python
X = [[1], [2], [3], [4], [5]]
Y = [[2], [4], [6], [8], [10]]
```

```python
w = random.random()
b = random.random()
```

```python
lr = 0.01
```

```python
for epoch in range(1000):
  y_pred = [[w * x[0] + b] for x in X]

  errors = [[y_pred[i][0] - Y[i][0]] for i in range(len(Y))]
  avg_error = sum([errors[i][0] for i in range(len(errors))]) / len(errors)

  w_grad = sum([errors[i][0] * X[i][0] for i in range(len(X))]) / len(X)
  b_grad = sum([errors[i][0] for i in range(len(X))]) / len(X)

  w -= lr * w_grad
  b -= lr * b_grad
```

```python
print("weights:", w)
print("bias:", b)
```

```python
plt.scatter(X, Y, color = 'blue')
```

```python
y_pred = [w * x[0] + b for x in X]
plt.plot(X, y_pred, color = 'red')
plt.scatter(X, Y, color = 'blue')
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Neural Network Visualization")
```

# Program 14B

```python
import random
import math
```

```python
X = [[1], [2], [3], [4], [5]]
Y = [[2], [4], [6], [8], [10]]
```

```python
w = random.random()
b = random.random()
```

```python
lr = 0.01
```

```python
def binary_sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
def bipolar_sigmoid(x):
  return (2 / (1 + math.exp(-x)) - 1)
```

```python
for epoch in range(1000):
  y_pred = [[binary_sigmoid(w * x[0] + b)] for x in X]

  errors = [[y_pred[i][0] - Y[i][0]] for i in range(len(Y))]
  avg_error = sum([errors[i][0] for i in range(len(errors))]) / len(errors)

  w_grad = sum([errors[i][0] * X[i][0] * y_pred[i][0] * (1 - y_pred[i][0]) for i in range(len(X))]) / len(X)
  b_grad = sum([errors[i][0] * y_pred[i][0] * (1 - y_pred[i][0]) for i in range(len(X))]) / len(X)

  w -= lr * w_grad
  b -= lr * b_grad
```

```python
print("weights:", w)
print("bias:", b)
```

```python
y_pred = [w * x[0] + b for x in X]
plt.plot(X, y_pred, color = 'red')
plt.scatter(X, Y, color = 'blue')
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Neural Network Visualization")
```
