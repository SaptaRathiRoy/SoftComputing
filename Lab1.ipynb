# Experiment 1: Implement Union, Intersection, Complement and Difference operations on fuzzy sets. Also create fuzzy relation by Cartesian product of any two fuzzy sets and perform max-min composition on any two fuzzy relations.

```python
def fuzzy_union(set1, set2):
  union = []

  for i in range(len(set1)):
    union.append(max(set1[i], set2[i]))
  
  return union
```

```python
def fuzzy_intersection(set1, set2):
  intersection = []

  for i in range(len(set1)):
    intersection.append(min(set1[i], set2[i]))
  
  return intersection
```

```python
def fuzzy_complement(set1):
  complement = []

  for i in range(len(set1)):
    complement.append(1 - set1[i])
  
  return complement
```

```python
def fuzzy_difference(set1, set2):
  difference = []

  for i in range(len(set1)):
    difference.append(max(set1[i]-set2[i], 0))
  
  return difference
```

```python
A = [0.7, 0.4, 0.6]
B = [0.2, 0.6, 0.3, 0.9]
```

```python
union = fuzzy_union(A, B)
intersection = fuzzy_intersection(A, B)
complement = fuzzy_complement(A)
difference = fuzzy_difference(A, B)
```

```python
print(f"Union: {union}")
print(f"Intersection: {intersection}")
print(f"Complement: {complement}")
print(f"Difference: {difference}")
```

# Experiment 2: Create fuzzy relation by Cartesian product of any two fuzzy sets and perform max-min composition on any two fuzzy relations.

```python
def fuzzy_cartesion_product(set1, set2):
  relation = []
  for a in set1:
    row = []
    for b in set2:
      row.append(min(a, b))
    relation.append(row)
  
  return relation
```

```python
def fuzzy_min_max_composition(relation1, relation2):
  composition = []
  for i in range(len(relation1)):
    row = []
    for j in range(len(relation2[0])):
      max_min = 0
      for k in range(len(relation1[0])):
        max_min = max(max_min, min(relation1[i][k], relation2[k][j]))
      row.append(max_min)
    composition.append(row)
  
  return composition
```

```python
A = [0.7, 0.4, 0.6]
B = [0.2, 0.6, 0.3, 0.9]
```

```python
relation1 = fuzzy_cartesion_product(A, B)
relation2 = fuzzy_cartesion_product(B, A)
```

```python
composition = fuzzy_min_max_composition(relation1, relation2)
```

```python
print(f"Relation 1: {relation1}")
print(f"Relation 2: {relation2}")
print(f"Composition: {composition}")
```
