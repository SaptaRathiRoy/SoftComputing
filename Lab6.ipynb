```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
input_size = 2
hidden_size = 3
output_size = 1
```

```python
W1 = np.random.randn(input_size, hidden_size)
W2 = np.random.randn(hidden_size, output_size)
```

```python
learning_rate = 0.1
epochs = 10000
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
losses = []
for epoch in range(epochs):
  z1 = np.dot(X, W1)
  a1 = sigmoid(z1)
  z2 = np.dot(a1, W2)
  y_pred = sigmoid(z2)

  loss = np.mean((y - y_pred)**2)
  if epoch % 1000 == 0:
    print(f"Epoch {epoch}: Loss: {loss:.4f}")
  losses.append(loss)
  
  delta2 = (y - y_pred) * sigmoid(z2) * (1 - sigmoid(z2))
  delta1 = delta2.dot(W2.T) * sigmoid(z1) * (1 - sigmoid(z1))
  dW2 = a1.T.dot(delta2)
  dW1 = X.T.dot(delta1)

  W2 += learning_rate * dW2
  W1 += learning_rate * dW1

# print("Final weights:")
# print("W1:", W1)
# print("W2:", W2)
```

```python
plt.plot(losses)
```

```python
import matplotlib.pyplot as plt
```

```python
def threshold(x, theta):
  if x >= theta:
    return 1
  else:
    return 0

def and_neuron(x1, x2):
  w1, w2 = 1, 1
  y = threshold(w1*x1 + w2*x2, 2)
  return y

def not_neuron(x):
  w = -1
  y = threshold(w*x, -0.5)
  return y
```

```python
and_inputs = [(0, 0), (0, 1), (1, 0), (1, 1)]
and_outputs = [and_neuron(x[0], x[1]) for x in and_inputs]

not_inputs = [0, 1]
not_outputs = [not_neuron(x) for x in not_inputs]
```

```python
plt.subplot(2, 2, 1)
plt.title("AND Function")
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter([x[0] for x in and_inputs], [x[1] for x in and_inputs], c = and_outputs)

plt.subplot(2, 2, 2)
plt.title("AND Function Outputs")
plt.xlabel('Inputs')
plt.ylabel('Outputs')
plt.scatter(range(len(and_outputs)), and_outputs)
```

```python
plt.subplot(2, 2, 1)
plt.title("Not Function")
plt.xlabel('x1')
plt.ylabel('x2')
plt.scatter(not_inputs, not_outputs, c = not_outputs)

plt.subplot(2, 2, 2)
plt.title("Not Function Outputs")
plt.xlabel('Inputs')
plt.ylabel('Outputs')
plt.scatter(range(len(not_outputs)), not_outputs)
```
