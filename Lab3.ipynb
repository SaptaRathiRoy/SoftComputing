# Experiment 6: Discrete Discourse and Continuous Discourse for unrelated fuzzy sets along with visualization.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
discrete_set = [0, 0, 1, 1, 1, 0, 0]
x = np.arange(len(discrete_set))
```

```python
print(list(zip(*enumerate(discrete_set))))
```

```python
# plt.stem(x, discrete_set, '-.')
plt.stem(*zip(*enumerate(discrete_set)), '-.')
plt.title('Discrete Fuzzy Set')
plt.ylabel('Membership Grade')
plt.xlabel('Universe of Discourse')
plt.show()
```

```python
mean = 5
std_dev = 2
x = np.linspace(0, 10, 100)
continuous_set =np.exp(-(x - mean)**2 / 2 * std_dev**2)
```

```python
plt.plot(x, continuous_set)
plt.title('Continuous Fuzzy Set')
plt.ylabel('Membership Grade')
plt.xlabel('Universe of Discourse')
plt.show()
```

# Experiment 7: Gaussian Distribution from fuzzy set to plot the distribution density using Bell-shaped curve.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mu = 0
sigma = 1

x = np.linspace(-5, 5, 100)
pdf = (1/(sigma*np.sqrt(2*np.pi)))*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
plt.figure(figsize = (8, 6))
plt.xlim(-5, 5)
plt.ylim(0, 0.5)
plt.plot(x, pdf, 'r', linewidth = 2)
plt.xlabel('X (Discourse)')
plt.ylabel('Probability')
plt.title('Gaussian Distribution')
```

```python
one_sigma = mu + sigma * np.array([-1, 1])
two_sigma = mu + sigma * np.array([-2, 2])
three_sigma = mu + sigma * np.array([-3, 3])
```

```python
plt.fill_between(x, pdf, where = (x > one_sigma[0]) & (x < one_sigma[1]), color = 'b', alpha = 0.4)
plt.fill_between(x, pdf, where = (x > two_sigma[0]) & (x < two_sigma[1]), color = 'g', alpha = 0.4)
plt.fill_between(x, pdf, where = (x > three_sigma[0]) & (x < three_sigma[1]), color = 'y', alpha = 0.4)
plt.show()
```

# Experiment 8: Implement Support Vector Machine classification by fuzzy concepts.

```python
import numpy as np
from sklearn import svm
```

```python
X = np.array([[0.7, 0.4], [0.6, 0.5], [0.8, 0.7], [0.5, 0.3], [0.4, 0.6]])
y = np.array([1, 0, 1, 0, 1])
```

```python
clf = svm.SVC(kernel = 'linear', C = 1.0)
clf.fit(X, y)
```

```python
x_new = np.array([[0.9, 0.5]])
prediction = clf.predict(x_new)
print(f"Prediction: {prediction}")
```
